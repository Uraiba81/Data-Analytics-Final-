## ***17. Replace income value [<=50k with 0 and >50k with 1]***

```python
df['income'] = df['income'].apply(lambda x: 0 if x == '<=50K' else 1)

df.income
```

## ***18. Which Workclass getting The Highest Salary***

```python
average_salary = df.groupby('workclass')['income'].mean()

highest_salary_workclass = average_salary.idxmax()
highest_salary_value = average_salary.max()
highest_salary_workclass
```

## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python
total_count = df['gender'].value_counts()
high_income_count = df[df['income'] == 1]['gender'].value_counts()
proportion = (high_income_count / total_count) * 100

print("Proportion of males earning >50K:", proportion.get('Male', 0), "%")
print("Proportion of females earning >50K:", proportion.get('Female', 0), "%")
```
